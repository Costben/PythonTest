```python
import pandas as pd
import numpy as np

# import CSV files

orders = pd.read_csv('20621098.csv')
customer  = pd.read_csv('customer.csv')
geolocation= pd.read_csv('geolocation.csv')
order_item= pd.read_csv('order_item.csv')
order_payment= pd.read_csv('order_payment.csv')
order_review= pd.read_csv('order_review.csv')
product_category_name_translation= pd.read_csv('product_category_name_translation.csv')
product= pd.read_csv('product.csv')
seller= pd.read_csv('seller.csv')
```

```python
#Check NaN data in each files
for i in [orders,customer,geolocation,order_item,order_payment,order_review,product_category_name_translation,product,seller]:
    print(i.isnull().sum(),'\n\n\n')
```

```python
#Drop all NaN, except order_review(Lack of comment)
orders  = orders.dropna()
customer  = customer.dropna()
geolocation  = geolocation.dropna()
product  = product.dropna()
order_item  = order_item.dropna()
order_payment  = order_payment.dropna()
product_category_name_translation  = product_category_name_translation.dropna()
seller  = seller.dropna()

data_1 = [orders,customer,geolocation,order_item,order_payment,product_category_name_translation,seller,product]
```

```python
for i in [orders,customer,geolocation,order_item,order_payment,order_review,product_category_name_translation,product,seller]:
    print(i.describe(),'\n')
```

```python
# Use Z-score to find and cope the outliner in numeric data

data_1 = [orders,customer,geolocation,order_item,order_payment,product_category_name_translation,seller,product]

import numpy as np
from scipy import stats
for index, df in enumerate(data_1):
    num_cols = df.select_dtypes(exclude=['object']).columns
    z = np.abs(stats.zscore(df[num_cols]))
    data_1[index] = df[(z < 3).all(axis=1)]

```

```python
#Convert Product Date to int
num_product_cols = product.select_dtypes(exclude=['object']).columns
product[num_product_cols]= product[num_product_cols].astype(int)
```

```python
# Check again
for i in data_1:
    print(i.isnull().sum(),'\n')
```
